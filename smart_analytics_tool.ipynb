```python
 import pandas as pd
import numpy as np
import plotly.express as px
from google.colab import files
```

```python
uploaded=files.upload()
if len(uploaded)!=1:
  print("Please upload only one dataset")
else:
  file=list(uploaded.keys())[0]
  df=pd.read_csv(file)
  print("Dataset loaded successfully")
```

```python
print("MISSING VALUES ANALYSIS \n")
missingValues=df.isnull().sum()
missingValues
```

```python
fig=px.bar(x=missingValues.index,y=missingValues.values,title="Missing values analysis",labels={"x":"Columns","y":"missing values"})
fig.show()
```

```python
print("STATISTICAL SUMMARY")
df.describe()
```

```python
numeric_columns=df.select_dtypes(include=np.number).columns.tolist()
numeric_columns
```

```python
import ipywidgets as widgets
```

```python
hist_column=widgets.Dropdown(options=numeric_columns,description="Column:")
hist_column
```

```python
fig=px.histogram(df,x=hist_column.value,title=f"histogram of {hist_column.value}")
fig.show()
```

```python
box_column=widgets.Dropdown(options=numeric_columns,description="Column:")
box_column
```

```python
fig=px.box(df,x=box_column.value,title=f"box plot of {box_column.value}")
fig.show()
```

```python
x_column=widgets.Dropdown(options=numeric_columns,description="X axis: ")
x_column
```

```python
y_column=widgets.Dropdown(options=numeric_columns,description="y axis: ")
y_column
```

```python
fig=px.scatter(df,x=x_column.value,y=y_column.value,title=f"{x_column.value} vs {y_column.value}")
fig.show()
```

```python
correlation=df.select_dtypes(include=np.number).corr()
correlation
```

```python
fig=px.imshow(correlation,text_auto=True,title="Correlation matrix")
fig.show()
```
